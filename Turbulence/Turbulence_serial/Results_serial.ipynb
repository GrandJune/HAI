```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
legend_properties = {'weight':'bold'}
data_folder = r"E:\data\hai-0520\Turbulence_5\Turbulence_alpha_turbulence_0.5_8"

organic_performance_file = data_folder + r"\organic_performance"
organic_knowledge_file = data_folder + r"\organic_knowledge"
organic_steps_file = data_folder + r"\organic_steps"
organic_knowledge_quality_file = data_folder + r"\organic_knowledge_quality"

pair_performance_file = data_folder + r"\pair_performance"
pair_knowledge_file = data_folder + r"\pair_knowledge"
pair_steps_file = data_folder + r"\pair_steps"
pair_knowledge_quality_file = data_folder + r"\pair_knowledge_quality"

with open(organic_performance_file, 'rb') as infile:
    organic_performance_across_episodes = pickle.load(infile)
with open(organic_knowledge_file, 'rb') as infile:
    organic_knowledge_across_episodes = pickle.load(infile)
with open(organic_steps_file, 'rb') as infile:
    organic_steps_across_episodes = pickle.load(infile)
with open(organic_knowledge_quality_file, 'rb') as infile:
    organic_knowledge_quality_across_episodes = pickle.load(infile)

with open(pair_performance_file, 'rb') as infile:
    pair_performance_across_episodes = pickle.load(infile)
with open(pair_knowledge_file, 'rb') as infile:
    pair_knowledge_across_episodes = pickle.load(infile)
with open(pair_steps_file, 'rb') as infile:
    pair_steps_across_episodes = pickle.load(infile)
with open(pair_knowledge_quality_file, 'rb') as infile:
    pair_knowledge_quality_across_episodes = pickle.load(infile)

organic_knowledge_quality_across_episodes = np.mean(organic_knowledge_quality_across_episodes, axis=0)
pair_knowledge_quality_across_episodes = np.mean(pair_knowledge_quality_across_episodes, axis=0)

# organic_knowledge_quality_across_episodes = np.array(organic_knowledge_quality_across_episodes)
# print(np.shape(organic_knowledge_quality_across_episodes))
# pair_knowledge_quality_across_episodes = np.array(pair_knowledge_quality_across_episodes)
# print(np.shape(pair_knowledge_quality_across_episodes))
```

```python
# Performance
# x = [50, 100, 150, 200, 250, 300, 350, 400, 450]
x = range(len(pair_performance_across_episodes))
plt.plot(x, pair_performance_across_episodes, "k:", label="Hybrid")
plt.plot(x, organic_performance_across_episodes, "k-", label="Organic")
plt.xlabel('Learning Length', fontweight='bold', fontsize=12)
plt.ylabel('Informativeness', fontweight='bold', fontsize=12)
# plt.xticks(x)
# ax.set_ylim(0.50, 0.95)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\performance_under_turbulence.png", transparent=True, dpi=300)
plt.show()
plt.clf()
# print(pair_knowledge_across_episodes)
```

```python
# Steps
# x = [50, 100, 150, 200, 250, 300, 350, 400, 450]
x = range(len(pair_performance_across_episodes))
plt.plot(x, pair_steps_across_episodes, "k:", label="Hybrid")
plt.plot(x, organic_steps_across_episodes, "k-", label="Organic")
plt.xlabel('Learning Length', fontweight='bold', fontsize=12)
plt.ylabel('Steps', fontweight='bold', fontsize=12)
# plt.xticks(x)
# ax.set_ylim(0.50, 0.95)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\steps_under_turbulence.png", transparent=True, dpi=300)
plt.show()
plt.clf()
# print(pair_knowledge_across_episodes)
```

```python
# Knowledge I
# x = [50, 100, 150, 200, 250, 300, 350, 400, 450]
x = range(len(organic_performance_across_episodes))
plt.plot(x, pair_knowledge_quality_across_episodes, "k:", label="Hybrid")
plt.plot(x, organic_knowledge_quality_across_episodes, "k-", label="Organic")
plt.xlabel('Learning Length', fontweight='bold', fontsize=12)
plt.ylabel('Representation Quality', fontweight='bold', fontsize=12)
# plt.xticks(x)
# ax.set_ylim(0.50, 0.95)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\knowledge_quality_under_turbulence.png", transparent=True, dpi=300)
plt.show()
plt.clf()
# print(pair_knowledge_across_episodes)
```
