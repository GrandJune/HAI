```python
import matplotlib.pyplot as plt
import pickle
legend_properties = {'weight':'bold'}
data_folder = r"E:\data\hai-0520\Strategies_3\AdaptiveTrust"

file = data_folder + r"\valence_across_episodes"


with open(file, 'rb') as infile:
    data = pickle.load(infile)

# Performance
x = list(range(0, 200, 10))
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, data, "k-s", label="Optimal Trust")
plt.xlabel('Learning Length', fontweight='bold', fontsize=12)
plt.ylabel('Optimal Trust', fontweight='bold', fontsize=12)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\Trust_across_episodes.png", transparent=True, dpi=300)
plt.show()
plt.clf()
# should be start with 0.73 (organic) and 0.82 (pair)
# at the third sample, should be as close as to 1% difference
```
